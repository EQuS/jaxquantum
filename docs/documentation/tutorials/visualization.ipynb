This notebook demonstrates how to utilize `plot_wigner` and `plot_qfunc` functions for visualizing quantum states, including single and batched states such as coherent and GKP states.

```python
import jaxquantum as jqt
import jaxquantum.codes as jqtb
import matplotlib.pyplot as plt
import jax.numpy as jnp
import numpy as np
```

## Visualizing the Wigner Function for Coherent States

```python
# Parameters for phase space grid
xvec = jnp.linspace(-3, 3, 51)
yvec = jnp.linspace(-3, 3, 51)

# Create a coherent state |α⟩ with amplitude α = 1.0
alpha = 1.0
state = jqt.displace(30, alpha) @ jqt.basis(30, 0)

# Plot the Wigner function
jqt.plot_wigner(state, xvec, yvec)
plt.show()
```

## Batched Visualization of Wigner Functions for Coherent States

```python
# Multiple displacements (batch of coherent states)
alphas = jnp.array([0.5, 1.0, 1.5])
states = jqt.displace(30, alphas) @ jqt.basis(30, 0)

# Plot the batch of Wigner functions with subtitles
jqt.plot_wigner(states, xvec, yvec, subtitles=np.array([f"α = {alpha}" for alpha in alphas]))
plt.show()
```

## Visualizing the Q-Function for GKP States

```python
# Generate a GKP state (logical |+⟩ state)
gkp_qubit = jqtb.GKPQubit({"N": 30, "delta": 0.4})
state = gkp_qubit.basis["+z"]

# Plot the Q-function
jqt.plot_qfunc(state, xvec, yvec)
plt.show()
```
