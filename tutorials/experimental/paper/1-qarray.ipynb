```python
import jaxquantum as jqt
import jax.numpy as jnp
import matplotlib.pyplot as plt
```

```python
Ncav = 10; Nq = 2; π=jnp.pi;
ωq = 2*π*1; ωc = 2*π*2; g=2*π*0.1;
a = jqt.create(Ncav) ^ jqt.identity(Nq)
σz = jqt.identity(Ncav) ^ jqt.sigmaz()
n = a.dag()*a; H = ωc*n + ωq*σz + g*n*σz
si = jqt.coherent(Ncav,1) ^ jqt.basis(Nq,1)
ts = jnp.linspace(0, 2, 101)
sf = jqt.sesolve(si, ts, H0=H)
at = jqt.calc_expect(a, sf)

fig, ax = plt.subplots(1,1, dpi=200, figsize=(4,3))
ax.plot(ts, at, "o", color="purple", alpha=0.1, markersize=5)
```
