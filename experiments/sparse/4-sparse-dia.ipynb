```python
import jaxquantum as jqt

from jax import jit, vmap
import jax.numpy as jnp
```

```python
def test_func(alpha):
    N = 30
    coherent_state = jqt.coherent(N, alpha)

    a = jqt.destroy(N).to_sparse_dia()

    return jqt.overlap(coherent_state, a.dag() @ a)
```

```python
test_func(2.5)
```

```python
na = 15 # Hilbert space dimension

a = jqt.destroy(na).to_sparse_dia() # annihilaiton operator
psi0 = jqt.basis(na, 0) # initial state

H = 0.0 # zero Hamiltonian
kappa_2 = 1 # two-photon loss rate
alpha = 2.0 # cat size
tsave = jnp.linspace(0, 3, 100)

loss_op = jnp.sqrt(kappa_2)*(a @ a - alpha**2 * jqt.qeye(na).to_sparse_dia()) # loss operator
loss_op = jqt.Qarray.from_list([loss_op]) # convert to list of loss operators

res = jqt.mesolve(H, psi0, tsave, c_ops=loss_op)
```
