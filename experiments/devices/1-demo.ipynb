```python
import jaxquantum as jqt
import jaxquantum.devices as jqd
import jax.numpy as jnp
```

```python
Ec_q = 0.3
Ej_q = 15

qubit = jqd.Transmon.create(
    23,
    {"Ec": Ec_q, "Ej": Ej_q},
    N_pre_diag=23,
)

res = jqd.Resonator.create(
    23,
    {"Ec": Ec_q, "El": Ej_q},
    N_pre_diag=23,
    basis=jqd.BasisTypes.fock,
    hamiltonian=jqd.HamiltonianTypes.full,
)
```

```python
phi_vals = jnp.linspace(-0.5,0.5,500)
axs = qubit.plot_wavefunctions(phi_vals, max_n = 7, mode="real")
```
