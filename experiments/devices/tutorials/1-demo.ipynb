# 4 - auto differentiation / metric

```python
@jit
def calculate_metrics(f_r):
    # devices ---

    # Resonator
    _, Ec_a, El_a = jqtd.calculate_lambda_over_four_resonator_zpf(f_r, 50)

    resonator = jqtd.Resonator.create(
        10,
        {"Ec": Ec_a, "El": El_a},
        N_pre_diag=10,
    )

    # Fluxonium
    Ec_q = 1
    El_q = 0.5
    Ej_q = 8

    qubit = jqtd.Fluxonium.create(
        25,
        {"Ec": Ec_q, "El": El_q, "Ej": Ej_q, "phi_ext": 0.47},
        use_linear=False,
        N_pre_diag=100,
    )

    # system ----
    g_rq = 0.3

    devices = [resonator, qubit]
    r_indx = 0
    q_indx = 1
    Ns = [device.N for device in devices]

    a0 = jqtd.promote(resonator.ops["a"], r_indx, Ns)
    a0_dag = jqtd.promote(resonator.ops["a_dag"], r_indx, Ns)

    q0 = jqtd.promote(qubit.ops["a"], q_indx, Ns)
    q0_dag = jqtd.promote(qubit.ops["a_dag"], q_indx, Ns)

    couplings = []
    couplings.append(-g_rq * (a0 - a0_dag) @ (q0 - q0_dag))

    system = jqtd.System.create(devices, couplings=couplings)
    system.params["g_rq"] = g_rq

    # metrics ----
    Es, kets = system.calculate_eig()

    # chi
    χ_e = Es[1:, 1] - Es[:-1, 1]
    χ_g = Es[1:, 0] - Es[:-1, 0]
    χ = χ_e - χ_g

    # kerr
    # kerr[0,n] = (E(n+2, g) - E(n+1, g)) - (E(n+1, g) - E(n, g))
    # kerr[1,n] = (E(n+2, e) - E(n+1, e)) - (E(n+1, e) - E(n, e))
    K_g = (Es[2:, 0] - Es[1:-1, 0]) - (Es[1:-1, 0] - Es[0:-2, 0])
    K_e = (Es[2:, 1] - Es[1:-1, 1]) - (Es[1:-1, 1] - Es[0:-2, 1])

    χ, K_g, K_e

    K_avg = (K_g + K_e)/2

    return jnp.abs(χ[0]/K_avg[0])

grad_calculate_metrics = jit(grad(calculate_metrics))
```

```python
grad_calculate_metrics(4.0)
```

# 5 - auto differentiation / noise sensitivity

```python
@jit
def calculate_resonator_freq(phi_ext):
    # devices ---

    # Resonator
    _, Ec_a, El_a = jqtd.calculate_lambda_over_four_resonator_zpf(4.0, 50)

    resonator = jqtd.Resonator.create(
        10,
        {"Ec": Ec_a, "El": El_a},
        N_pre_diag=10,
    )

    # Fluxonium
    Ec_q = 1; El_q = 0.5; Ej_q = 8

    qubit = jqtd.Fluxonium.create(
        25,
        {"Ec": Ec_q, "El": El_q, "Ej": Ej_q, "phi_ext": phi_ext},
        use_linear=False,
        N_pre_diag=100,
    )

    # region: system operators ----
    g_rq = 0.3

    devices = [resonator, qubit]; r_indx = 0; q_indx = 1
    Ns = [device.N for device in devices]

    a0 = jqtd.promote(resonator.ops["a"], r_indx, Ns)
    a0_dag = jqtd.promote(resonator.ops["a_dag"], r_indx, Ns)

    q0 = jqtd.promote(qubit.ops["a"], q_indx, Ns)
    q0_dag = jqtd.promote(qubit.ops["a_dag"], q_indx, Ns)
    # endregion ---

    couplings = []
    couplings.append(-g_rq * (a0 - a0_dag) @ (q0 - q0_dag))

    system = jqtd.System.create(devices, couplings=couplings)
    system.params["g_rq"] = g_rq

    # metrics ----
    Es, kets = system.calculate_eig()

    omega_r = Es[1,0] - Es[0,0]
    return omega_r 

grad_calculate_resonator_freq = jit(grad(calculate_resonator_freq))
```

```python
calculate_resonator_freq(0.5)
```

```python
flux_values = jnp.linspace(0,1,101)
domega_dphi = 2 * jnp.pi * jit(vmap(grad_calculate_resonator_freq))(flux_values)
sqrt_A_phi = 10e-6 # Phi_0
gamma_phi = sqrt_A_phi * jnp.sqrt(jnp.log(2)) * jnp.abs(domega_dphi)
T_phi = 1 / gamma_phi
```

```python

fig, ax = plt.subplots(1,1,dpi=200, figsize=(6,4))

ax.plot(flux_values, T_phi / 1e6)
ax.set_ylabel(r"Storage Dephasing Limit $\Gamma_{\Phi}^{-1}$ (ms)")
ax.set_xlabel("External Flux (Phi_0)")
ax.set_yscale("log")
ax.set_title(r"$\Gamma_{\Phi} = \left|\frac{d\omega_r}{d\Phi_{\rm ext}}\right|\sqrt{A_{\Phi} \ln{2}}$")
ax.set_ylim(0,1e2)
ax.axhline(1, color="k", ls="--", lw=0.5)
```
